```python
import numpy as np
import matplotlib.path as mpath
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt


points = bezier.input_points("Test_1")
Bezier = bezier.special_Bezier_interpolation(points)

Bezier_etalon = bezier.input_points("Bezier_1_etalon")


Path = mpath.Path

fig, (ax1, ax2) = plt.subplots(1, 2)
fig.set_figheight(15)
fig.set_figwidth(5)

pcode = [Path.MOVETO, Path.CURVE4, Path.CURVE4, Path.CURVE4]

N = np.shape(points)[0] - 1

for ic in range(N):
    pp1 = mpatches.PathPatch(Path(Bezier[4 * ic : 4 * ic + 4], pcode),
                             fc="none")
    ax1.add_patch(pp1)
ax1.plot()

for ic in range(N):
    pp2 = mpatches.PathPatch(Path(Bezier_etalon[4 * ic : 4 * ic + 4], pcode),
                             fc="none", ec="red")
    ax2.add_patch(pp2)
ax2.plot()

ax1.set_title('Solution')
ax2.set_title('Etalon')


plt.show()
```
